```python
import jieba
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# 根据实际情况纠正jieba分词的不准确
#jieba.load_userdict('dict_digitaleconomy.txt')
```

```python
stopword_file = 'E:/Tools/stopwords/baidu_stopwords.txt'
process_file = 'E:/Tools/En_Text_Split_Wordcloud/text_digital_economy.txt'
```

```python
#停用词表
with open(stopword_file, 'r',encoding='UTF-8') as f:
    stopword_list = [word.strip('\n') for word in f.readlines()]
    print(stopword_list)
```

```python
# 分词，并去掉停用词
result = ""
with open(process_file,encoding = "utf-8") as f:
    mytext = f.read()
word_list = jieba.lcut(mytext)
for w in word_list:
    if w not in stopword_list:
        result = result + " " + w
```

```python
print(result)
```

```python
# 统计词频
df = pd.DataFrame(result.split( ),columns =["words"])
res_count = df.groupby(["words"]).size().sort_values(ascending=False)
res_count.to_excel("result_count.xlsx")
```

```python
# 词云图
wordcloud = WordCloud().generate(result)
plt.imshow(wordcloud, interpolation = "bilinear")
plt.axis("off")
```
